```python
# Importing Libraries
from ast import literal_eval
import pandas as pd
import json
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
# Importing Data
df1 = pd.read_json('receipts.json',lines=True)
df2 = pd.read_json('brands.json',lines=True)
df3 = pd.read_json('users.json',lines=True)
```

Cleaning receipts data

```python
# Normalizing nested column rewardsReceiptItemList
df1 = df1.explode('rewardsReceiptItemList')   
df1.reset_index(inplace=True)
df1_norm = pd.json_normalize(df1['rewardsReceiptItemList'],errors='ignore')
df1 = pd.merge(df1, df1_norm, left_index=True, right_index=True, how='outer')
df1.drop(['rewardsReceiptItemList', 'index'], axis=1, inplace=True)
df1
```

```python
# Function to convert dates
def date_converter(x):
    if pd.isnull(x):
        return None
    try:
        return(datetime.utcfromtimestamp(int(x)/1000).strftime('%Y-%m-%d %H:%M:%S'))
    except TypeError:
        return(None)
```

```python
# Cleaning columns
df1['_id'] = df1['_id'].apply(lambda x: x['$oid'] if isinstance(x, dict) and '$oid' in x else x)
df1['createDate'] = df1['createDate'].apply(lambda x: x['$date'] if isinstance(x, dict) and '$date' in x else x)
df1['dateScanned'] = df1['dateScanned'].apply(lambda x: x['$date'] if isinstance(x, dict) and '$date' in x else x)
df1['finishedDate'] = df1['finishedDate'].apply(lambda x: x['$date'] if isinstance(x, dict) and '$date' in x else x)
df1['modifyDate'] = df1['modifyDate'].apply(lambda x: x['$date'] if isinstance(x, dict) and '$date' in x else x)
df1['pointsAwardedDate'] = df1['pointsAwardedDate'].apply(lambda x: x['$date'] if isinstance(x, dict) and '$date' in x else x)
df1['purchaseDate'] = df1['purchaseDate'].apply(lambda x: x['$date'] if isinstance(x, dict) and '$date' in x else x)
```

```python
# Converting dates
df1['createDate'] = df1['createDate'].apply(lambda x: date_converter(x))
df1['dateScanned'] = df1['dateScanned'].apply(lambda x: date_converter(x))
df1['finishedDate'] = df1['finishedDate'].apply(lambda x: date_converter(x))
df1['modifyDate'] = df1['modifyDate'].apply(lambda x: date_converter(x))
df1['pointsAwardedDate'] = df1['pointsAwardedDate'].apply(lambda x: date_converter(x))
df1['purchaseDate'] = df1['purchaseDate'].apply(lambda x: date_converter(x))
df1
```

Cleaning Brands data

```python
df2_norm = pd.json_normalize(df2['cpg'])
df2_norm= df2_norm.add_prefix('cpg.')
df2['_id'] = df2['_id'].apply(lambda x: x['$oid'] if isinstance(x, dict) and '$oid' in x else x)
df2.drop(['cpg'], axis=1, inplace=True)
df2 = pd.merge(df2, df2_norm, left_index=True, right_index=True, how='outer')
df2['topBrand'] = df2['topBrand'].astype('bool')
df2
```

Cleaning users data table

```python
df3['_id'] = df3['_id'].apply(lambda x: x['$oid'] if isinstance(x, dict) and '$oid' in x else x)
df3['createdDate'] = df3['createdDate'].apply(lambda x: x['$date'] if isinstance(x, dict) and '$date' in x else x)
df3['lastLogin'] = df3['lastLogin'].apply(lambda x: x['$date'] if isinstance(x, dict) and '$date' in x else x)
df3['createdDate'] = df3['createdDate'].apply(lambda x: date_converter(x))
df3['lastLogin'] = df3['lastLogin'].apply(lambda x: date_converter(x))
df3
```

```python
# Visualizing account creations over time
df3['createdDate'] = pd.to_datetime(df3['createdDate'])
df3['month_year'] = df3['createdDate'].dt.to_period('M')
creation_counts = df3.groupby('month_year').size()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(creation_counts.index.astype(str), creation_counts.values)
plt.xlabel('Month-Year')
plt.ylabel('Number of User Account Creations')
plt.title('Number of User Account Creations Over Time')
plt.show()
```

We can see a sharp increase in the number of user account created in 12/2020, in comparison to the previous month. This could an anomaly or data of user accounts from previous months could be missing.
